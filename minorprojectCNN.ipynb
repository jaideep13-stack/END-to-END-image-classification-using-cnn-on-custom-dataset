```python
# CNNC for image classification
# here we use the CIFAR-10 datasets for cnn
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.datasets import CIFAR10
```

```python
# dataset and dataloader
from torch.utils.data import DataLoader
import torchvision.transforms as transforms # this is transform the data into tensor and scale the data

transform = transforms.Compose([
    transforms.ToTensor(),# this function helps to covert image into tensor data
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])# these are the standard values we can use


# next thing we want dataset or traning set 
# CIFARO have already the traning and testing datasets

trainset= CIFAR10(root = "./all_dataset",train = True,download = True,transform = transform)
testset = CIFAR10(root ="./all_dataset",train = False,download = True, transform= transform)            

```

```python
trainset
```

```python
trainloader = DataLoader(trainset,batch_size = 64,shuffle = True)
testloader = DataLoader(testset,batch_size=64)
# why we choose batch size 64 beacuse the is big the data is small we chosse 32 or very big data we chosse 128 
```

```python
# build cnn (multi_class classification)
# in crossentropy loss ===> softmax activation function
class CNN (nn.Module):
    def __init__(self):
        super(CNN,self). __init__()

        self.Conv_layers = nn.Sequential(
            nn.Conv2d(3,32,kernel_size = 3,padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),# kernel_size=2,stide = 2


            nn.Conv2d(32,64,kernel_size = 3,padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),# kernel_size=2,stide = 2


            nn.Conv2d(64,128,kernel_size = 3,padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2,2)# kernel_size=2,stide = 2

        )

        self.fc_layers = nn.Sequential(
            nn.Linear(4*4*128,256),
            nn.ReLU(),

            nn.Linear(256,10)

        )

    def forward (self,x):
        x = self.Conv_layers(x)
        x= x.view(x.size(0),-1)# flattering
        x= self.fc_layers(x)

        return x

        
            
            

            
            
            
            

        
```

```python
model = CNN()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
# tranning of cnn
epochs = 10

for epoch in range (epochs):
    epoch_tranning_loss = 0.0

    for images,labels in trainloader :
        optimizer.zero_grad()

        output = model.forward(images)# forward propagation
        loss = criterion(output,labels)# loos fnx
        loss.backward()# backward propagation
        optimizer.step()# update params


        epoch_tranning_loss+=loss.item()

    print(f"eoch = {epoch+1}/{epochs}loss ={epoch_tranning_loss/len(trainloader)}")    
```

```python
# evaluate our cnn
correct_labels = 0
total_labels = 0

model.eval()

with torch.no_grad():
    for images,labels in testloader:
        output = model.forward(images)
        _, predicted = torch.max(output,1)

        correct_labels+=(predicted==labels).sum().item()
        total_labels+=labels.size(0)
    print (f"accuracy ={correct_labels/total_labels*100}")    
        
```
